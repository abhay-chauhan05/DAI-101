**Medium Problem**


**Write code to find top 5 employees with highest percent_basepay in each year having overtimepay non zero.** (percent_basepay=basepay/totalpay *100)

**reference:** https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.nlargest.html

```python
data['percent_basepay'] = (data['BasePay'] / data['TotalPay']) * 100
overpay_data = data[data['OvertimePay'] != 0]

top5_emp = overpay_data.groupby('Year').apply(lambda x: x.nlargest(5, 'percent_basepay'))[['EmployeeName', 'percent_basepay']]
top5_emp=top5_emp.reset_index()
top5_emp = top5_emp.rename(columns={'level_1': 'Id'})[['Id', 'EmployeeName', 'Year', 'percent_basepay']]

display(top5_emp)
```
